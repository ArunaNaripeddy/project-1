```python
# data frame to hold randomly generated latitude/longitude coords to will eventually go into response
loc_data = pd.DataFrame()
loc_data['rand_lat'] = [np.random.uniform(-90,90) for x in range(1500)]
loc_data['rand_long'] = [np.random.uniform(-180, 180) for x in range(1500)]
# 
loc_data['closest_city'] = ""
loc_data['country'] = ""
# loop to find the closest city 
for index, row in loc_data.iterrows():
    lat = row['rand_lat']
    long = row['rand_long']
    loc_data.set_value(index, 'closest_city', citipy.nearest_city(lat,long).city_name)
    loc_data.set_value(index, 'country', citipy.nearest_city(lat,long).country_code)
```

```python
base_url = "https://api.breezometer.com/baqi/?"
counter = 0 
for index, row in loc_data.iterrows():
    city = row['closest_city']
    country = row['country']
    print("Getting coords for city: %s: %s, %s" %(index, city, country))
    target_url = '%slat=%s&lon=%s&key=%s&fields=breezometer_aqi,random_recommendations,dominant_pollutant_canonical_name,pollutants,country_color,datetime,dominant_pollutant_text,dominant_pollutant_description,country_aqi' %(base_url, lat, long, bm_key)
    print(target_url)
    counter += 1
print(counter)    
```
